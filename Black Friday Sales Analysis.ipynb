Now, after we have cleaned our data and gotten enough insights, depending on the problem you are solving you can use machine models to help you make better decisions with predictions you get.


Not all data related problems need machine learning. Some only need visualization and you get what you want. An example of this is when you want a summary of the kinds of people living in an environment or a community. Data visualization will do just fine here and you get your answer. 


Now, when you have a particular data about a group of schools in a community with details of each student. After doing a lot of analysis, you find out that the ratio of boys to girls is 3:1. 


Now, given another data about groups of schools in another community and you weren't told the number of boys or girls but you were only given the details of each student, you cna create a model that trains on the the dataset you have initially with labels of boys and girls and make predictions using this new dataset we have.


For the dataset, we have the data of customers, the products they purchased and the total amount of purchase. Now, this purchase amount is our label but its not a classification problem and so we cannot group people's purchases into classes because it's continuous(not fixed). Instead, is a regression problem. 


There are various classification and regresion models. I only listed a few in this notebook. You can go and do more research about other kinds.

### NB: The fact that a model works best for a particular problem does not guarrante that it will always work for all kinds of problems. You keep trying out models and comparing them to find the best.

There are various evaluation metrics that we can use such as accuracy, f1-score, roc-auc, rmse, log_loss, etc. Depending on the type of problem we are solving.


For the dataset we will use rmse - root mean squared error.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
lre  = lr.predict(x_test)

from sklearn.metrics import mean_squared_error
from math import sqrt
sqrt(mean_squared_error(y_test, lre))
```

```python
# BayesianRidge Regression
from sklearn import linear_model
clf = linear_model.BayesianRidge()
clf.fit(x_train, y_train)
pred  = clf.predict(x_test)

from sklearn.metrics import mean_squared_error
from math import sqrt
sqrt(mean_squared_error(y_test, pred))
```

```python
# Linear regression model that is robust to outliers
# HuberRegressor
from sklearn.linear_model import HuberRegressor

hubr = HuberRegressor()
hubr.fit(x_train, y_train)
pred  = hubr.predict(x_test)

from sklearn.metrics import mean_squared_error
from math import sqrt
sqrt(mean_squared_error(y_test, pred))
```

```python
from sklearn.linear_model import ElasticNet

regr = ElasticNet(random_state=0)
regr.fit(x_train, y_train)
pred  = regr.predict(x_test)

from sklearn.metrics import mean_squared_error
from math import sqrt
sqrt(mean_squared_error(y_test, pred))
```

```python
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor(max_depth=6, random_state=0)
tree.fit(x_train, y_train)
pred  = tree.predict(x_test)

from sklearn.metrics import mean_squared_error
from math import sqrt
sqrt(mean_squared_error(y_test, pred))
```

An extremely randomized tree regressor.


Extra-trees differ from classic decision trees in the way they are built.
When looking for the best split to separate the samples of a node into two
groups, random splits are drawn for each of the `max_features` randomly
selected features and the best split among those is chosen. When
`max_features` is set 1, this amounts to building a totally random
decision tree.
Warning: Extra-trees should only be used within ensemble methods.

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import ExtraTreeRegressor

extra_tree = ExtraTreeRegressor(random_state=0)
reg = BaggingRegressor(extra_tree, random_state=0).fit(x_train, y_train)
pred  = reg.predict(x_test)

from sklearn.metrics import mean_squared_error
from math import sqrt
sqrt(mean_squared_error(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(x_train, y_train)
pred = rf.predict(x_test)

from sklearn.metrics import mean_squared_error
from math import sqrt

sqrt(mean_squared_error(pred, y_test))
```

```python
from lightgbm import LGBMRegressor

lgbm = LGBMRegressor(n_estimators=700, learning_rate=0.3)
lgbm.fit(x_train, y_train)
a = lgbm.predict(x_test)

from sklearn.metrics import mean_squared_error
from math import sqrt

sqrt(mean_squared_error(a, y_test))
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(estimator=LGBMRegressor(n_estimators=700, learning_rate=0.3), X=X, y=y,scoring="neg_mean_squared_error", cv=10)

for i in cv_scores:
    print(sqrt(abs(i)))
print(sqrt(abs(cv_scores).mean()))
print(sqrt(abs(cv_scores).std()))
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(random_state=0, n_estimators=500)
xgb.fit(x_train, y_train)
b = xgb.predict(x_test)
sqrt(mean_squared_error(b, y_test))
```

```python
from catboost import CatBoostRegressor
from catboost import MetricVisualizer

from sklearn.model_selection import train_test_split

data = train_test_split(X, y, test_size=0.2, random_state=0)
X_train, X_validation, y_train, y_validation = data


from catboost import Pool
train_pool = Pool(
    data=X_train, 
    label=y_train, 
)

validation_pool = Pool(
    data=X_validation, 
    label=y_validation, 
)

model1 = CatBoostRegressor(iterations=1000,
                           learning_rate=0.2,

                           )
model1.fit(
    train_pool,
    eval_set=validation_pool,
    verbose=False,
    plot=True
)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

b = model1.predict(x_test)
sqrt(mean_squared_error(b, y_test))
```
